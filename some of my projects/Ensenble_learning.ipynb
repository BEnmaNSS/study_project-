```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.3)

```

```python
X_train = df_train.drop(["class"], axis=1)
y_train = df_train["class"]

X_test = df_test.drop(["class"],axis=1)
y_test = df_test["class"]

print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

# choosing our hyperparameter

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
import time
```

```python
# Setting hyperparameters for Gradient Boosting model
hyperparameters_gb = {
    'learning_rate': [0.1, 0.15],
    'n_estimators': [10, 50, 100],
    'max_depth': [3, 5]
    }

# Initializing the Gradient Boosting model
model_gb_tuned = GradientBoostingClassifier()
start_time = time.time()
# Performing randomized search cross validation
clf_gb_tuned = RandomizedSearchCV(model_gb_tuned, hyperparameters_gb, scoring='accuracy', n_jobs=-1, n_iter=2)

# Fitting the tuned classifier on the training data
clf_gb_tuned.fit(X_train, y_train)
end_time = time.time()



# Printing the validation score and the best parameters of the model
print(f"The validation score for the tuned Gradient Boosting model is {clf_gb_tuned.best_score_}")
print(f"The best parameters of the tuned Gradient Boosting model are {clf_gb_tuned.best_params_}")
```

```python
# Printing the execution time
# Printing the execution time
execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")
```

```python
# Setting hyperparameters for Random Forest
hyperparameters_rf = {
    'n_estimators': [60,70],
    'max_depth': [4,5,6,9],
    'min_samples_split': [2, 3,4],
    'min_samples_leaf': [ 3,4]
}


# Initializing the Gradient Boosting model
model_rf = RandomForestClassifier()

# Performing randomized search cross validation
clf_rf_tuned = RandomizedSearchCV(model_rf, hyperparameters_rf, scoring='accuracy', n_jobs=-1, n_iter=5)

# Fitting the tuned classifier on the training data
clf_rf_tuned.fit(X_train, y_train)

# Printing the validation score and the best parameters of the model
print(f"The validation score for the tuned Random Forest is {clf_rf_tuned.best_score_}")
print(f"The best parameters of the tuned Random Forest model are {clf_rf_tuned.best_params_}")
```

```python
#setting hyperparameter for svm
hyperparameters_svm = {
    'C': (0.1, 10),  # Continuous uniform distribution between 0.1 and 10
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'gamma': (0.01, 1),  # Continuous uniform distribution between 0.01 and 1
    'degree': (2, 6),  # Discrete uniform distribution between 2 and 5
    'class_weight': [None, 'balanced']
}
# sv model
model_svm = SVC()
#
clf_svm_tuned = RandomizedSearchCV(model_svm, hyperparameters_svm , scoring='accuracy', n_jobs=-1, n_iter=5)

# Fitting the tuned classifier on the training data
clf_svm_tuned.fit(X_train, y_train)

# Printing the validation score and the best parameters of the model
print(f"The validation score for the tuned svm Forest is {clf_svm_tuned.best_score_}")
print(f"The best parameters of the tuned svm Forest model are {clf_svm_tuned.best_params_}")

```
